```python
# 07.04.2013 - 
# Comment: With 5-point moving average algorithm 
#
# I decided to clean my code a little bit so that it's easier for you to see the overall flow!
# Let me know if you have any questions

# Importing software packages
import pandas as pd
from scipy import stats as stats
from pandas import DataFrame, read_csv

# Change the below location as needed!
Location = r'C:\Users\MoonSoo Choi\Desktop\Summer_2013\Brainpage\example_out.csv'
df = pd.read_csv(Location)

import numpy as np
import scipy.signal as signal
import scipy as sp
import scipy.fftpack as fourier
from pylab import *
import matplotlib.pyplot as plt

# Checking whether the csv file is imported correctly
df

# Extracting x-, y-, z- axis columns
x_accel = df.loc[:, ['x_accel']]
y_accel = df.loc[:, ['y_accel']]
z_accel = df.loc[:, ['z_accel']]

Xravel = ravel(x_accel)
Yravel = ravel(y_accel)
Zravel = ravel(z_accel)

# This is a moving-average function
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    return (ret[n - 1:] - ret[:1 - n]) / n

XMA_5 = moving_average(Xravel, n=5)
YMA_5 = moving_average(Yravel, n=5)
ZMA_5 = moving_average(Zravel, n=5)

# This is an arbitrary value to distinguish between a step and a sedantary moment of a Spire user
step_still_threshold = 5

# Declaring various array-type variables
# Each array has a size of 4695, because k-point moving average truncates the last k points

# Legend
# old = old_fixed
# new = new_fixed (new_fixed is updated iff there is a significant amount of change in the acceleration; otherwise, if the
#                  acceleration change is negligible, new_fixed will not be updated and will stay the same)

# 'sampleold' and 'samplenew' variables represent 'old_fixed' and 'new_fixed' variables for x-,y-,z- 3-axis magnitude 
# xyz_magnitude = sqrt(x**2 + y**2 + z**2)

sampleold = np.arange(0,4695,dtype=np.float)
samplenew = np.arange(0,4695,dtype=np.float)

sampleoldX = np.arange(0,4695,dtype=np.float)
samplenewX = np.arange(0,4695,dtype=np.float)

sampleoldY = np.arange(0,4695,dtype=np.float)
samplenewY = np.arange(0,4695,dtype=np.float)

sampleoldZ = np.arange(0,4695,dtype=np.float)
samplenewZ = np.arange(0,4695,dtype=np.float)

# This algorithm measures max, min, and threshold values every 0.5 seconds (5 samples)
# threshold = (max+min)/2
XMAnew_5_maxf = np.arange(0,4695,dtype=np.float)
XMAnew_5_minf = np.arange(0,4695,dtype=np.float)
XMAnew_5_thresholdf = np.arange(0,4695,dtype=np.float)

YMAnew_5_maxf = np.arange(0,4695,dtype=np.float)
YMAnew_5_minf = np.arange(0,4695,dtype=np.float)
YMAnew_5_thresholdf = np.arange(0,4695,dtype=np.float)

ZMAnew_5_maxf = np.arange(0,4695,dtype=np.float)
ZMAnew_5_minf = np.arange(0,4695,dtype=np.float)
ZMAnew_5_thresholdf = np.arange(0,4695,dtype=np.float)

# All data points identified as "step candidates" are stored in step_temp_array before final step validation stage
step_temp_array = ravel(np.zeros(4700,dtype=np.float))

# Each data point will be categorized into 3 different groups; step, unclear, and still
step_array = ravel(np.zeros(4700,dtype=np.float))
unclear_array = ravel(np.zeros(4700,dtype=np.float))
still_array = ravel(np.zeros(4700,dtype=np.float))

# changeXYZ is an indicator array --
#    1 if one of the three axes show a significant acceleration change
#    0 if otherwise
changeXYZ = np.arange(0,4695,dtype=np.float)

# Variable Initiazliation
sampleoldX[0] = XMA_5[0]
samplenewX[0] = XMA_5[0]

sampleoldY[0] = YMA_5[0]
samplenewY[0] = YMA_5[0]

sampleoldZ[0] = ZMA_5[0]
samplenewZ[0] = ZMA_5[0]

for i in xrange(1, 4695):
    changeX = 0
    changeY = 0
    changeZ = 0
    
    # If there is a significant amount of change in each axis
    if np.abs(XMA_5[i] - samplenewX[i-1]) > step_still_threshold and np.abs(XMA_5[i] - XMA_5[i-1]) > step_still_threshold:
        changeX = 1

    if np.abs(YMA_5[i] - samplenewY[i-1]) > step_still_threshold and np.abs(YMA_5[i] - YMA_5[i-1]) > step_still_threshold:
        changeY = 1
        
    if np.abs(XMA_5[i] - samplenewZ[i-1]) > step_still_threshold and np.abs(ZMA_5[i] - ZMA_5[i-1]) > step_still_threshold:
        changeZ = 1
    
    # If at least one of the axes show a significant change
    if changeX + changeY + changeZ > 0:
        changeXYZ[i] = 1
        samplenewX[i] = XMA_5[i-1]
        sampleoldX[i] = samplenewX[i-1]
        
        samplenewY[i] = YMA_5[i-1]
        sampleoldY[i] = samplenewY[i-1]
        
        samplenewZ[i] = ZMA_5[i-1]
        sampleoldZ[i] = samplenewZ[i-1]
    
    # If none of the axes show a significant change
    elif changeX + changeY + changeZ == 0:
        changeXYZ[i] = 0
        sampleoldX[i] = samplenewX[i-1]
        samplenewX[i] = samplenewX[i-1]
        
        sampleoldY[i] = samplenewY[i-1]
        samplenewY[i] = samplenewY[i-1]
        
        sampleoldZ[i] = samplenewZ[i-1]
        samplenewZ[i] = samplenewZ[i-1]

step_temp_index_old = 0
step_temp_index_new = 0
step = 0

# Calculating max, min, and threshold for every 5 samples
for n in xrange(0, 4695):
    XMAnew_5_maxf[n] = np.amax(samplenewX[5*(n/5):5*(n/5+1)])
    XMAnew_5_minf[n] = np.amin(samplenewX[5*(n/5):5*(n/5+1)])
    XMAnew_5_thresholdf[n] = np.float((XMAnew_5_maxf[n] + XMAnew_5_minf[n])/2)
    
    YMAnew_5_maxf[n] = np.amax(samplenewY[5*(n/5):5*(n/5+1)])
    YMAnew_5_minf[n] = np.amin(samplenewY[5*(n/5):5*(n/5+1)])
    YMAnew_5_thresholdf[n] = np.float((YMAnew_5_maxf[n] + YMAnew_5_minf[n])/2)
    
    ZMAnew_5_maxf[n] = np.amax(samplenewZ[5*(n/5):5*(n/5+1)])
    ZMAnew_5_minf[n] = np.amin(samplenewZ[5*(n/5):5*(n/5+1)])
    ZMAnew_5_thresholdf[n] = np.float((ZMAnew_5_maxf[n] + ZMAnew_5_minf[n])/2)
    
# Calculating max-min difference for each axis
    Xdiff = XMAnew_5_maxf[n] - XMAnew_5_minf[n]
    Ydiff = YMAnew_5_maxf[n] - YMAnew_5_minf[n]
    Zdiff = ZMAnew_5_maxf[n] - ZMAnew_5_minf[n]
     
for n in xrange(1, 4695):
       
    # Below -- (Just observing X-axis)
    #    if samplenewX[n] - samplenewX[n-1] < 0 and samplenewX[n-1] > XMAnew_5_thresholdf[n] and samplenewX[n] < XMAnew_5_thresholdf[n]:
    # Below -- (Just observing Y-axis)
    #    if (samplenewY[n] - samplenewY[n-1] < 0 and samplenewY[n-1] > YMAnew_5_thresholdf[n] and samplenewY[n] < YMAnew_5_thresholdf[n]):
    
    if (samplenewZ[n] - samplenewZ[n-1] < 0 and samplenewZ[n-1] > ZMAnew_5_thresholdf[n] and samplenewZ[n] < ZMAnew_5_thresholdf[n]): 
        step_temp_array[n] = 1
        step_temp_index_old = step_temp_index_new
        step_temp_index_new = n
        #print step_temp_index_old
     
        # This is when the step is unclear
        if np.sum(step_temp_array[step_temp_index_old:step_temp_index_old+75]) < 3 and np.sum(step_temp_array[step_temp_index_old-75:step_temp_index_old-1]) < 3 or \
            step_temp_index_new - step_temp_index_old == 1:
            unclear_array[step_temp_index_old] = 1
            
        # This is when the person either begins or is currently in the stepping process
        elif np.sum(step_temp_array[step_temp_index_old:step_temp_index_old+50]) >= 2 and \
            np.sum(step_temp_array[step_temp_index_old:step_temp_index_old+25]) >= 1 and \
            step_temp_index_new - step_temp_index_old > 1:
                    
            step_array[step_temp_index_old] = 1
            step = step + 1
            
        # This is when the step ends
        elif np.sum(step_temp_array[step_temp_index_old:step_temp_index_old+25]) <= 1 and \
            np.sum(step_temp_array[step_temp_index_old - 25:step_temp_index_old]) > 1 and \
            step_temp_index_new - step_temp_index_old > 1:
                
            step_array[step_temp_index_old] = 1
            step = step + 1
        
        # This is when the person is still/sedantary
        else:
            still_array[step_temp_index_old] = 1
    
    # This is when the person is still/sedantary
    else: 
        step_temp_array[n] = 0
        still_array[n] = 1
        
beginning = 0
afterwards = 5000

plt.gcf().set_size_inches(10, 10)
plt.subplot(221)
plt.plot(dummy_x,step_array[beginning:beginning+afterwards],dummy_x,unclear_array[beginning:beginning+afterwards],'-',dummy_x,still_array[beginning:beginning+afterwards],'.')
plt.legend(['Steps','Unclear','Still'])
plt.title('Comparing overall movements')

plt.xlim(0,5000)
plt.ylim(0,2)
print 'Number of steps'
print np.sum(step_array[1:4695])

print 'Number of unclear steps'
print np.sum(unclear_array[1:4695])


#plt.figure(2)
plt.subplot(222)
plt.plot(samplenewX[beginning:beginning+afterwards])
plt.plot(XMAnew_5_maxf[beginning:beginning+afterwards])
plt.plot(XMAnew_5_minf[beginning:beginning+afterwards])
plt.plot(XMAnew_5_thresholdf[beginning:beginning+afterwards])
plt.title('filtered X axis')

#plt.figure(3)
plt.subplot(223)
plt.plot(samplenewY[beginning:beginning+afterwards])
plt.plot(YMAnew_5_maxf[beginning:beginning+afterwards])
plt.plot(YMAnew_5_minf[beginning:beginning+afterwards])
plt.plot(YMAnew_5_thresholdf[beginning:beginning+afterwards])
plt.title('filtered Y axis')

#plt.figure(4)
plt.subplot(224)
plt.plot(samplenewZ[beginning:beginning+afterwards])
plt.plot(ZMAnew_5_maxf[beginning:beginning+afterwards])
plt.plot(ZMAnew_5_minf[beginning:beginning+afterwards])
plt.plot(ZMAnew_5_thresholdf[beginning:beginning+afterwards])
plt.title('filtered Z axis')
```
